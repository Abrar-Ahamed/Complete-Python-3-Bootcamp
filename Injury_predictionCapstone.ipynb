```python
def replace_numbering(s):
  # Define the regular expression pattern
  pattern = r"^\d+\. " # Matches a number followed by a dot and a space at the beginning of a line

  # Replace the pattern with an empty string
  replaced_text = re.sub(pattern, "", s, flags=re.MULTILINE)

  return replaced_text
```
